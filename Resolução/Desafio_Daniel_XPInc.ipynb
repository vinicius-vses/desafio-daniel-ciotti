Escreva um script que:

```python
import pandas as pd
from datetime import datetime
from IPython.display import HTML

dados = {
    'DATA': ['03/03/2020', '03/03/2020', '03/03/2020', '03/03/2020', '03/03/2020',
             '04/03/2020', '04/03/2020', '04/03/2020', '04/03/2020', '04/03/2020', '04/03/2020',
             '05/03/2020', '05/03/2020', '05/03/2020', '05/03/2020', '05/03/2020', '05/03/2020'],
    'ATIVO': ['DI1F21', 'DI1J21', 'DI1N21', 'DI1F22', 'DI1F25',
              'DI1F21', 'DI1J21', 'DI1N21', 'DI1F22', 'DI1F25', 'DI1F27',
              'DI1F21', 'DI1J21', 'DI1N21', 'DI1F22', 'DI1F25', 'DI1F29'],
    'QTD': [10500, 3500, 2435, 4545, 180,
            9500, 3500, 2100, 4800, 180, 330,
            8000, 3300, 2050, 6300, 250, 1110]
}

def datetime_to_excel(date_str):
    date = datetime.strptime(date_str, '%d/%m/%Y')
    excel_start_date = datetime(1900, 1, 1)
    delta = date - excel_start_date
    return delta.days + 1

df = pd.DataFrame(dados)
df['DATA_EXCEL'] = df['DATA'].apply(datetime_to_excel)

styled_df = df.style.set_properties(**{
    'border': '1px solid black',
    'padding': '8px',
    'text-align': 'center',
    'font-family': 'Courier New, monospace',
    'font-size': '14px'
}).set_table_styles([
    {
        'selector': 'th',
        'props': [
            ('background-color', '#003366'),
            ('color', 'white'),
            ('border', '1px solid black'),
            ('text-align', 'center'),
            ('padding', '8px')
        ]
    }
])

styled_df.set_table_attributes("style='border-collapse:collapse'").hide(axis='index')
```

**Exercício 2:**

Crie um script que resolva os seguinte exercício:

3. Tratar as informações para gerar a tabela que indica quanto cada cliente tem de financeiro em ações.

```python
import pandas as pd

df['Financeiro em ações'] = df['QTD'] * df['Preco']

df_financeiro = df.groupby('Cliente')['Financeiro em ações'].sum().reset_index()

df_financeiro['Financeiro em ações'] = df_financeiro['Financeiro em ações'].apply(lambda x: f'R${x:,.2f}')

def style_table(df):
    return df.style.set_table_styles(
        [
            {'selector': 'thead th', 'props': [('border', '1px solid black'), ('text-align', 'center')]},
            {'selector': 'tbody td', 'props': [('border', '1px solid black'), ('text-align', 'center')]},
        ]
    ).hide(axis='index')

style_table(df_financeiro)
```
